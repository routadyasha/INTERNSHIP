```python
try:
    url=driver.current_url
    print('URL_EXTRACTED',url)
    lat_lan=re.findall(r'@(.*)data',url)
    if len(lat_lan):
        lat_lan_list = lat_lan[0].split(',')
        if len(lat_lan_list)>=2:
            lat = lat_lan_list[0]
            lan=lat_lan_list[1]
        print('Latititude={},Longitude={}'.format(lat,lan))
except Exception as e:
    print("Error:",str(e))
```
